```python
def getposition(letter):
    lis = []
    for i in range(len(res)):
        if res[i]==letter:
            lis.append(i+1)
            
    diflis = []
#     for i in range(len(lis)-1):
#         diflis.append(lis[i+1]-lis[i])
        
    for i in range(0, len(lis)-1):
        for j in range(1, len(lis)):
            if i!=j:
                diflis.append(abs(lis[j] - lis[i]))
           
    return diflis
```
