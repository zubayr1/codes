```python
def getposition(letter):
    lis = []
    for i in range(len(data)):
        if data[i]==letter:
            lis.append(i+1)
            
    diflis = []

        
    for i in range(0, len(lis)-1):
        for j in range(1, len(lis)):
            if i!=j:
                diflis.append(abs(lis[j] - lis[i]))
           
    return diflis
```

```python
def getall():
    pos=[]
    for i in set(data):
        pos+=getposition(i)
    dic={}
    for i in pos:
        if i in dic:
            dic[i]+=1
        else:
            dic[i]=1
    return pos, dic
```

```python
def makechunks(length):
    
    lis = [{},{},{},{},{},{},{},{}]
    
    for i in range(len(data)):
        if data[i] in lis[i%length]:
            lis[i%length][data[i]]+=1
        else:
            lis[i%length][data[i]]=1
        
                   
    
    return lis


def makechunksbigram(length):
    
    lis = [{},{},{},{},{},{},{},{}]
    
    for i in range(len(data)-1):
        if (data[i] + data[i+1]) in lis[i%length]:
            lis[i%length][(data[i] + data[i+1])]+=1
        else:
            lis[i%length][(data[i] + data[i+1])]=1
        
                   
    
    return lis
```

```python
ALPHABET = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
def decrypt(ciphertext,key):
    plaintext = ''
    for i in range(len(ciphertext)):
        p = ALPHABET.index(ciphertext[i])
        k = ALPHABET.index(key[i%len(key)])
        c = (p - k) % 26
        plaintext += ALPHABET[c]
    return plaintext
```

```python
def index_of_coincidence(text):
    counts = [0]*26
    for char in text:
        counts[ALPHABET.index(char)] += 1
    numer = 0
    total = 0
    for i in range(26):
        numer += counts[i]*(counts[i]-1)
        total += counts[i]
    return 26*numer / (total*(total-1))

found = False
period = 0
while not found:
    period += 1
    slices = ['']*period
    for i in range(len(data)):
        slices[i%period] += data[i]
    sum = 0
    for i in range(period):
        sum += index_of_coincidence(slices[i])
    ioc = sum / period
    if ioc > 1.6:
        found = True
        
period
```

```python
from math import sqrt
def cosangle(x,y):
    numerator = 0
    lengthx2 = 0
    lengthy2 = 0
    for i in range(len(x)):
        numerator += x[i]*y[i]
        lengthx2 += x[i]*x[i]
        lengthy2 += y[i]*y[i]
    return numerator / sqrt(lengthx2*lengthy2)
```
