```python
import hashlib

text = 's8bomini'

for i in range(1000, 9999):
    password = text + str(i)
    result = hashlib.md5(password.encode('utf-8')).hexdigest()
    
        
    if result=="17ddb7280fbf073bf56dc57380c50f71":
        print(text + str(i))
```

```python
def myfunc(st):
    str=''
    for index, l in enumerate(st):
        if index % 2 == 0:
            str+=l.upper()
        else:
            str+=l.lower()
    return str
```
